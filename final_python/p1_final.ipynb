```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the datasets from Excel files
df_hindi = pd.read_excel('cleaned_bollywood_movies.xlsx')
df_telugu = pd.read_excel('cleaned_tollywood_movies.xlsx')
df_indian = pd.read_excel('cleaned_indian_movies.xlsx')
```

```python
# Calculate ROI for each movie
# ROI = (BoxOffice - Budget) / Budget
df_hindi['ROI'] = (df_hindi['BoxOffice__Crores'] - df_hindi['Budget__Crores']) / df_hindi['Budget__Crores']
df_telugu['ROI'] = (df_telugu['BoxOffice__Crores'] - df_telugu['Budget__Crores']) / df_telugu['Budget__Crores']
df_indian['ROI'] = (df_indian['BoxOffice__Crores'] - df_indian['Budget__Crores']) / df_indian['Budget__Crores']
```

```python
# Round ROI to 2 decimal places for clarity
df_hindi['ROI'] = round(df_hindi['ROI'], 2)
df_telugu['ROI'] = round(df_telugu['ROI'], 2)
df_indian['ROI'] = round(df_indian['ROI'], 2)
```

```python
# Find top 3 Hindi movies by ROI
top_3_hindi = df_hindi[df_hindi['Language'] == 'Hindi'][['Title', 'ROI', 'BoxOffice__Crores', 'Budget__Crores']].sort_values('ROI', ascending=False).head(3)
print("Top 3 Hindi Movies by ROI:")
print(top_3_hindi)
print("\n")
```

```python
# Find top 3 Telugu movies by ROI
top_3_telugu = df_telugu[['Title', 'ROI', 'BoxOffice__Crores', 'Budget__Crores']].sort_values('ROI', ascending=False).head(3)
print("Top 3 Telugu Movies by ROI:")
print(top_3_telugu)
print("\n")
```

```python
# Combine all datasets to find top 3 movies across all industries
df_all = pd.concat([df_hindi, df_telugu, df_indian], ignore_index=True)
top_3_all = df_all[['Title', 'ROI', 'BoxOffice__Crores', 'Budget__Crores', 'Language']].sort_values('ROI', ascending=False).head(3)
print("Top 3 Movies Across All Industries by ROI:")
print(top_3_all)
print("\n")

#  Plotting the results
# Plot for top 3 Hindi movies
plt.figure(figsize=(8, 5))
plt.bar(top_3_hindi['Title'], top_3_hindi['ROI'], color='lightblue')
plt.title('Top 3 Hindi Movies by ROI')
plt.xlabel('Movie Title')
plt.ylabel('ROI')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
plt.savefig('top_3_hindi_roi.png')
plt.close()
```

```python
# Plot for top 3 Telugu movies
plt.figure(figsize=(8, 5))
plt.bar(top_3_telugu['Title'], top_3_telugu['ROI'], color='lightgreen')
plt.title('Top 3 Telugu Movies by ROI')
plt.xlabel('Movie Title')
plt.ylabel('ROI')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
plt.savefig('top_3_telugu_roi.png')
plt.close()
```

```python
# Plot for top 3 movies across all industries
plt.figure(figsize=(8, 5))
plt.bar(top_3_all['Title'], top_3_all['ROI'], color='lightcoral')
plt.title('Top 3 Movies Across All Industries by ROI')
plt.xlabel('Movie Title')
plt.ylabel('ROI')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
plt.savefig('top_3_all_roi.png')
plt.close()
```
