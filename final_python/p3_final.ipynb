```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the datasets from Excel files
df_hindi = pd.read_excel('cleaned_bollywood_movies.xlsx')
df_telugu = pd.read_excel('cleaned_tollywood_movies.xlsx')
df_indian = pd.read_excel('cleaned_indian_movies.xlsx')
```

```python
# Combine all datasets into one
df_all = pd.concat([df_hindi, df_telugu, df_indian], ignore_index=True)
```

```python
# Calculate average BoxOffice collections by genre for combined dataset
avg_genre_all = round(df_all.groupby('Genre')['BoxOffice__Crores'].mean(), 2).reset_index()
avg_genre_all = avg_genre_all.sort_values('BoxOffice__Crores', ascending=False)
```

```python
# Plot average BoxOffice collections by genre for combined dataset
plt.figure(figsize=(10, 8))
plt.barh(avg_genre_all['Genre'], avg_genre_all['BoxOffice__Crores'], color='skyblue')
plt.title('Average BoxOffice Collections by Genre (All Indian Movies)')
plt.xlabel('Average BoxOffice (Crores)')
plt.ylabel('Genre')
plt.tight_layout()
plt.savefig('avg_boxoffice_all_genres.png')
plt.show()
```

# Inferences for All Indian Movies:
- The highest-performing genres, such as Biographical Sports Drama and Epic Fantasy Action, have significantly higher average BoxOffice collections, likely due to their mass appeal and large-scale production.
- Genres like Sports Drama, Thriller, and Family Drama tend to have lower average collections, possibly due to niche audiences or less commercial appeal.
- The diversity of genres in the combined dataset reflects varied audience preferences across Indian cinema, with action, fantasy, and biographical themes dominating.

```python
# Step 6: Calculate average BoxOffice collections by genre for Bollywood (Hindi) movies
avg_genre_bolly = round(df_hindi[df_hindi['Language'] == 'Hindi'].groupby('Genre')['BoxOffice__Crores'].mean(), 2).reset_index()
avg_genre_bolly = avg_genre_bolly.sort_values('BoxOffice__Crores', ascending=False)
```

```python
# Step 7: Plot average BoxOffice collections by genre for Bollywood movies
plt.figure(figsize=(10, 8))
plt.barh(avg_genre_bolly['Genre'], avg_genre_bolly['BoxOffice__Crores'], color='lightcoral')
plt.title('Average BoxOffice Collections by Genre (Bollywood Movies)')
plt.xlabel('Average BoxOffice (Crores)')
plt.ylabel('Genre')
plt.tight_layout()
plt.savefig('avg_boxoffice_bollywood_genres.png')
plt.show()
```

# Inferences for Bollywood Movies:
- Bollywood has fewer genres compared to the combined dataset, indicating less variety in storytelling.
- Biographical Sports Drama remains a top performer, suggesting strong audience interest in real-life inspirational stories.
- Comedy, action, and romance genres perform relatively better in Bollywood than in the overall Indian cinema, reflecting Bollywood's focus on mainstream entertainment.
- Lower-performing genres like Family Drama and Thriller may be overshadowed by Bollywood's preference for historical, biographical, or action-oriented narratives.
