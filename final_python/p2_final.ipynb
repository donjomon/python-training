```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the datasets from Excel files
df_hindi = pd.read_excel('cleaned_bollywood_movies.xlsx')
df_telugu = pd.read_excel('cleaned_tollywood_movies.xlsx')
df_indian = pd.read_excel('cleaned_indian_movies.xlsx')
```

```python
#  Combine all datasets into one
df_all = pd.concat([df_hindi, df_telugu, df_indian], ignore_index=True)
```

```python
#  Calculate mean BoxOffice collections
mean_box_office = df_all['BoxOffice__Crores'].mean()
print(f"Mean BoxOffice Collections: {mean_box_office:.2f} Crores\n")
```

```python
# Plot histogram of BoxOffice collections
plt.figure(figsize=(8, 5))
df_all['BoxOffice__Crores'].plot(kind='hist', bins=10, edgecolor='black', color='skyblue')
plt.title('Distribution of BoxOffice Collections')
plt.xlabel('BoxOffice__Crores')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
#  Plot boxplot of BoxOffice collections
plt.figure(figsize=(8, 5))
plt.boxplot(df_all['BoxOffice__Crores'], vert=True, patch_artist=True, boxprops=dict(facecolor='lightgreen'))
plt.title('BoxOffice Collections Distribution')
plt.ylabel('BoxOffice__Crores')
plt.tight_layout()
plt.show()
```

```python
#  Identify outliers using IQR method
Q1 = df_all['BoxOffice__Crores'].quantile(0.25)
Q3 = df_all['BoxOffice__Crores'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
outliers = df_all[df_all['BoxOffice__Crores'] > upper_bound][['Title', 'BoxOffice__Crores', 'Language']]
print("Outliers in BoxOffice Collections (above upper bound):")
print(outliers)
print("\n")
```

# inferences 
- The histogram shows most movies earn less than 300 Crores, with a few earning much more, making the data lopsided.
- The average BoxOffice collection is about 322.21 Crores, but most movies earn less than this.
- The boxplot shows the middle range of earnings is around 100-300 Crores, with some movies earning way more.
- Outlier movies earn more than 591.25 Crores. These are the super hit films.
- These outliers are blockbuster movies that make a lot more money than most.
